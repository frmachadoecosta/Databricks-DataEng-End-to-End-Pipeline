```python
import dlt
from pyspark.sql.functions import *
```

## **DLT Pipeline**

```python
# Expectactions
my_rules = {
    "rule1": "product_id IS NOT NULL",
    "rule2": "product_name IS NOT NULL"
}

```

```python
@dlt.table(name="DimProducts_stage_v2")
@dlt.expect_all_or_drop(my_rules)
def DimProducts_stage():
  df = spark.readStream.table("databricks_cat.silver.products_silver")
  return df
```

**Streaming View**

```python
@dlt.view
def DimProducts_view_v2():
    df = spark.readStream.table("Live.DimProducts_stage_v2")
    return df
```

**DimProducts**

```python
dlt.create_streaming_table("DimProducts_v2")
```

```python
dlt.apply_changes(
    target = "DimProducts_v2",
    source = "DimProducts_view_v2",
    keys = ["product_id"],
    sequence_by = "product_id",
    stored_as_scd_type= 2
)
```
